```python
# Mapping old column names to new readable names
column_mapping = {
    'Age': 'AgeMean',  # Clarifying it's based on age mean of group
    'Sex': 'Gender',  # More descriptive than "Sex"
    'HighChol': 'HighCholesterol',
    'CholCheck': 'CholesterolCheck',
    'BMI': 'BodyMassIndex',
    'Smoker': 'SmokerStatus',
    'HeartDiseaseorAttack': 'HeartDiseaseHistory',
    'PhysActivity': 'PhysicalActivity',
    'Fruits': 'DailyFruitIntake',
    'Veggies': 'DailyVegetableIntake',
    'HvyAlcoholConsump': 'HeavyAlcoholConsumption',
    'GenHlth': 'GeneralHealth',
    'MentHlth': 'PoorMentalHealthDays',
    'PhysHlth': 'PoorPhysicalHealthDays',
    'DiffWalk': 'DifficultyWalking',
    'Stroke': 'StrokeHistory',
    'HighBP': 'HighBloodPressure',
    'Diabetes': 'DiabetesDiagnosis'
}

# Applying the new column names to the DataFrame
df.rename(columns=column_mapping, inplace=True)
```

```python
df.head()
```

```python
# Mapping age groups to mean values
age_group_to_mean = {
    1.0 : '18 to 24',
    2.0 : '25 to 29',
    3.0 : '30 to 34',
    4.0 : '35 to 39',
    5.0 : '40 to 44',
    6.0 : '45 to 49',
    7.0 : '50 to 54',
    8.0 : '55 to 59',
    9.0 : '60 to 64',
    10.0 : '65 to 69',
    11.0 : '70 to 74',
    12.0 : '75 to 79',
    13.0 : '80 to 100'
}

# Replace age groups with mean values
df['AgeMean'] = df['AgeMean'].map(age_group_to_mean)
```

```python
df.head()
```

```python
import re
# Function to calculate the mean of age ranges dynamically
def compute_age_mean(val):
    # Use regular expressions to extract numbers from the text
    match = re.findall(r'\d+', val)
    if match:  # If numbers are found
        lower_bound, upper_bound = map(int, match)
        return (lower_bound + upper_bound) / 2
    else:
        return None  # Return None for non-numerical groups

# Apply the function to the AGEG5YR column
df['AgeMean'] = df['AgeMean'].apply(compute_age_mean)

print(df)
```

```python
#line chart with the x axis being the age group and the y axis being the number of people in that age group
age_group_counts = df['AgeMean'].value_counts().sort_index()
age_group_counts.plot(kind='line', figsize=(10, 6), marker='o', color='b', linewidth=2)
plt.title('Number of People in Each Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of People')
```

```python
# bar chart for 60 to 70 age group, age as the x and y axis being mean of fruits and veggies
age_60_to_70 = df[df['AgeMean'].between(60, 70)]
age_60_to_70_fruit_veggie_means = age_60_to_70.groupby('AgeMean')[['DailyFruitIntake', 'DailyVegetableIntake']].mean()
age_60_to_70_fruit_veggie_means.plot(kind='bar', figsize=(10, 6))
plt.title('Mean Daily Fruit and Vegetable Intake for Ages 60 to 70')
plt.xlabel('Age')
plt.ylabel('Mean Intake')
```

```python
#all age groups bar chart, age as x and y axis being mean of fruits and veggies, overlay a line chart for mean of fruits and veggies
all_age_groups_fruit_veggie_means = df.groupby('AgeMean')[['DailyFruitIntake', 'DailyVegetableIntake']].mean()
all_age_groups_fruit_veggie_means.plot(kind='bar', figsize=(10, 6))
plt.title('Mean Daily Fruit and Vegetable Intake for All Age Groups')
plt.xlabel('Age')
plt.ylabel('Mean Intake')


```

```python
#bar chart where x axis is age and y axis being diabetes diagnosis
diabetes_diagnosis = df.groupby('AgeMean')['DiabetesDiagnosis'].mean()
diabetes_diagnosis.plot(kind='bar', figsize=(10, 6))
plt.title('Mean Diabetes Diagnosis Rate by Age')
plt.xlabel('Age')
plt.ylabel('Mean Diagnosis Rate')


```

```python
#bar chart where x axis is age and y being alcohol consumption and smoking 
alcohol_smoking = df.groupby('AgeMean')[['HeavyAlcoholConsumption', 'SmokerStatus']].mean()
alcohol_smoking.plot(kind='bar', figsize=(10, 6))
plt.title('Mean Alcohol Consumption and Smoking Rate by Age')
plt.xlabel('Age')
plt.ylabel('Mean Rate')



```

```python
#bar chart where x axis is smoking status and y axis being diabetes diagnosis
smoking_diabetes = df.groupby('SmokerStatus')['DiabetesDiagnosis'].mean()
smoking_diabetes.plot(kind='bar', figsize=(10, 6))
plt.title('Mean Diabetes Diagnosis Rate by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Mean Diagnosis Rate')
```

```python
#bar chart where x axis is alcohol status and y axis being diabetes diagnosis
alcohol_diabetes = df.groupby('HeavyAlcoholConsumption')['DiabetesDiagnosis'].mean()
alcohol_diabetes.plot(kind='bar', figsize=(10, 6))
plt.title('Mean Diabetes Diagnosis Rate by Alcohol Consumption Status')
plt.xlabel('Alcohol Consumption Status')
plt.ylabel('Mean Diagnosis Rate')


```
